```python
#fourth question


#part 1

import pickle

def createFile():
    file = open("book.dat","ab")
    BookNo = int(input("Book No:: "))
    Book_Name = input("Book Name:: ")
    Author =input("Author Name:: ")
    Price = int(input("PRICE:: "))
    record = [BookNo, Book_Name, Author, Price]
    pickle.dump(record, file)
    file.close()
    
#part 2

def countRec(Author):
    file = open("book.dat","rb")
    total = 0
    try:
        while True:
            record = pickle.load(file)
            if record[2]==Author:
                total+=1
    except EOFError:
        pass
    return total
    file.close()

```
